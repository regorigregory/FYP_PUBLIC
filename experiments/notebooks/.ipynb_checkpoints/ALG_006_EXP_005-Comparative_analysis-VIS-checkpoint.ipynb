<h2 id="Introduction">Introduction</h2>
<p>A more than incremental step upwards from constant weight (CW) approaches are adaptive weight approaches. In this notebook one of them, bilateral support weight windows were implemented to this pipeline.</p>
<h2 id="Abstract">Abstract</h2>
<p>Two variants of bilateral weights were &nbsp;introduced to the pipeline as part of the cost function. These were referred to as &ldquo;summed&rdquo; and &ldquo;product weight&rdquo; variants. As their name suggests, the summed entailed the summation of the intensity and spatial weights whereas the &ldquo;product&rdquo; expression entailed the multiplication of the aforementioned two components of bilateral weights. The achieved improvement with respect to the previous iteration, where CW windows were used, was 17%. Achieved percentage of bad pixels (with threshold of 1 &ndash; bad4 here) was 29.4 % which approximates the performance of the pipeline to be at around the 41<sup>st</sup> &nbsp;place in the Middlebury Stereo Evaluation Framework&rsquo;s (version 3) chart. Horizontal structural noise and a large number of streaking artefacts were observed upon visual analysis.</p>

<p>According to Fang, after CW, the proposed solutions branched into two categories: Cross-Based Aggregation (CROSS) and Adaptive Weight Aggregation (AW). While both ideas seemed interesting to consider for implementation, later work (Hamzah and Ibrahim, 2016)&nbsp; suggested that Adaptive Support Weights (ASW &ndash; which is equivalent to AW in Fang&rsquo;s terminology) would perform the best (Chen, Ardabilian and Chen, 2015).&nbsp;</p>
<p>Here, similarly to CW, cost is aggregated for both a&nbsp; reference and a target image pixel &nbsp;for a set of neighbouring pixels, though, their weight is dynamically calculated. A variety of methods has been proposed to calculate these dynamic weights (Hosni, Bleyer and Gelautz, 2013). One main &nbsp;characteristic of these various strategies that help differentiate between these methods is whether they performed in a symmetric or asymmetric manner. While the symmetric calculates a weight for both windows (left and right), asymmetric is only concerned with the left image window&rsquo;s weights. However, Hosni et al. concludes, there was no significant nor conclusive difference in accuracy between the two strategies in their comparison.</p>
<p>Perhaps one of the most advanced techniques to calculate adaptive weights is trilateral filtering (Chen, Ardabilian and Chen, 2015). Chen et al.&rsquo;s ASW technique was based on&nbsp; three main rules: colour, spatial distance and boundary. According to the colour rule, if the centre pixel and a pixel&nbsp; in the support window around it have similar intensities, the probability of them having the same disparities is high, therefore this pixel&rsquo;s weight in the window should be high. According to the&nbsp; spatial distance rule the greater the distance of a pixel from the centre pixel is, the lower the &nbsp;probability of them having the same disparity. The proposed boundary rule, which also happened to be the novelty in their solution,&nbsp; states that if there is a boundary (edge i.e. discontinuity)&nbsp; between two pixels then the probability of them having the same disparity is smaller. <p>However, &nbsp;as a first step towards adaptive weights in this project, this was deviated from in three regards. In this alternative implementation the first key difference was that&nbsp; only the spatial and colour terms were used as part of the cost aggregation function. Secondly, instead of using the support&nbsp; weight (w(p,q))&nbsp; term as the multiplicator of the cost matrix within a support window , it was made part of the cost calculation. This decision was made so as to be able to store intermediary weight results and not have to recalculate them thus speeding up stereo correspondence. &nbsp;Thirdly, instead of the sum of&nbsp; the two&nbsp; weight terms, their product was used.&nbsp;This was suggested by (De-Maeztu, Villanueva and Cabeza, 2011). <h2 id="Method">Method</h2>
<p>As it has been suggested above, two variants of the proposed bilateral filtering term was tested: one that used the &nbsp;product of the weight terms and one that used the sum of the weight terms.</p>
<p>First, in order to mitigate the effects of ill-parameter configuration and reduce the time needed to demonstrate relevant results, the two algorithm variants were tested on the Middlebury 2003 dataset&rsquo;s quarter resolution scenes in grayscale mode. The methodology was similar to the one followed for &nbsp;<a href="ALG_005_EXP_001-PatchMatch-2003.ipynb" target="_blank" rel="noopener">ALG_005_EXP_001</a>. However, here, first, a range of the parameters were tested with a relatively big step (20-50). Then, the ranges with the pest results were tested again, but with &nbsp;smaller steps (5-10). This way local minimums or locations near local minimums were though to have found. Notebook versions of these experiments can be found at <a href="ALG_006_EXP_001-Bilateral_product_2003.ipynb" target="_blank" rel="noopener">ALG_006_EXP_001-Bilateral_product_2003</a> and <a href="ALG_006_EXP_002-Bilateral_summed_2003.ipynb" target="_blank" rel="noopener">ALG_006_EXP_002-Bilateral_summed_2003</a>. Separate written analysis was not created for these experiments as it was assumed that their results would be misleading due to the limited size of the Middlebury 2003 dataset.</p>
<p>Therefore, informed by these two initial experiments, the Middlebury 2014 dataset&rsquo;s training images were tested at quarter resolution in grayscale for a set of parameters thought to be near the aforementioned local minimums. For the list of parameters please see Table 1.</p>

<p>Then, since significant difference was not identified between the summed and product weight approaches (<a href="ALG_006_EXP_003-Bilateral_product_2014-VIS.ipynb" target="_blank" rel="noopener">ALG_006_EXP_003-Bilateral_product_2014-VIS</a>, <a href="ALG_006_EXP_004-Bilateral_summed_2014-VIS.ipynb" target="_blank" rel="noopener">ALG_006_EXP_004-Bilateral_summed_2014-VIS</a>), a comparative analysis was conducted with regards to the experiments listed in Table 1.While specific objective of this project was to experiment on raw outputs of the algorithm, due to the large number of horizontal structural noise observed the disparity maps were additionally benchmarked after applying a median filter of 5 (width and height).</p>
<h2 id="Discussion-and-results">Discussion and results</h2>
<p>Compared to the results logged &nbsp;during the experiment on constant weight windows (CW) &nbsp;(ALG_005_EXP_001) there had been a 17% increase in accuracy (best CW: 46.01%, best bilateral: 29.4%) overall when analysing bad4 results.&nbsp;This would place the algorithm&rsquo;s performance at around the 41<sup>st</sup> &nbsp;place in the Middlebury Evaluation Framework&rsquo;s chart (Scharstein and Hirschm&uuml;ller, 2014) already overtaking some deep learning based solutions (DTS - (Bapat and Frahm, 2018), iResNet - (Liang <em>et al.</em>, 2017)).</p>
<p>Significant difference in terms of accuracy between the summed and product weight term methods was identified. However, &nbsp;the discrepancies in accuracy were thought to be on the account of more favourably selected and tested parameters for the summed version which manifested in a 4% difference overall. The greatest improvements in accuracy between the constant and bilateral weighted methods were &nbsp;achieved on the scenes Adirondack (27%) , ArtL (21%), MotorcycleE (51%) and PianoL (21%). This suggested that the application of bilateral weights (especially using the &ldquo;plus termed&rdquo; version) made the algorithm more robust towards lighting and exposure variations within a scene. This was an interesting result since this pipeline only considered pixel intensities, not gradients (which are thought to be more invariant when lighting and exposure conditions change (De-Maeztu, Villanueva and Cabeza, 2011)).&nbsp; The worst improvements were observed w.r.t. the scene Playtable. The reason behind this was thought to be the mainly diagonal orientation of edged within that scene which the algorithm was not able to interpret properly.</p>
<p>Informed by visual assessment, there was an increased amount of horizontal structural noise observable (Boyat and Joshi, 2015). This noise often manifested in a few pixel radius black pixel patch. Additionally, errors were thought to have manifested in streaking artefacts as well. &nbsp;The most severely affected scenes were MotorcycleE, Piano, PianoL, Pipes, Playroom and Playtable scenes.</p>
<p>Having observed the aforementioned noise, the best disparity maps were re-evaluated after applying a 5x5 median filter on them. The best improvement achieved was with regard to the &ldquo;product termed&rdquo; version, 3%, while the opposite effect was observable when it comes to the &ldquo;sum termed&rdquo; version. It was thought that should there had been a more favourable parameter setting for the &ldquo;product termed&rdquo; version, the observations would have aligned and a median filtering would have actually made its performance worse.</p>
<p>Please note, that experiments so far were performed on the training dataset and while the evaluation (test) dataset was thought to pose similar challenges, the validity of the comparisons to the Middlebury Evaluation Framework&rsquo;s 2014 version were not actually tested. Though, they were though to be good approximations.</p>
<h2 id="Conclusion">Conclusion</h2>
<p>There was no statistically relevant difference observed between the &ldquo;summed&rdquo; and the &ldquo;product&rdquo; variants. &nbsp;The achieved improvement with respect to the previous iteration, where constant weight (CW) windows were used, was 17%. &nbsp;Greatest improvements were achieved with regards to scenes of lighting or exposure variant. Horizontal structural noise and a large number of streaking artefacts were observed upon visual analysis. The "plus termed" version outperformed the "product termed" one, though, it was thought to be on the account on more favourable parameter selection.</p>

### Tesing the effect of applying a 5x5 median blur

```python
import cv2
import pandas as pd
import numpy as np
import project_helpers
import os
from components.utils.Metrix import Wrapper as me
from components.utils import utils as u

file_path = os.path.join(project_helpers.get_project_dir(), "benchmarking","MiddEval", "custom_log","best_results.csv" )

pd = pd.read_csv(file_path)

gt_mask = "disp0GT.pfm"
occ_mask = "mask0nocc.png"

average = dict(bm_30_9x3=0, blg_40_5x7gc_8_gs_90_alph_0=0, plusblg_45_5x7gc_5_gs_1_alph_0=0, bm_30_1x1=0)
average["trunc_plusblg_30_5x7no_infmge_30_-9_-1_gc_5_gs_1_a_30"] = 0
for img_path, scene, exp_id in pd[["image_filename", "scene", "experiment_id"]].values:

    root = project_helpers.get_project_dir()
    img_full_path = os.path.join(project_helpers.fix_win_rel_paths(img_path))
    img_raw = cv2.imread(img_full_path, cv2.IMREAD_GRAYSCALE)
    img_temp = cv2.medianBlur(img_raw, 5)
    dirname = os.path.dirname(img_full_path)
    
    gt_path = os.path.join(dirname, gt_mask)
    occ_path = os.path.join(dirname, occ_mask)
    
    
    gt = u.load_pfm(gt_path)[0]
    occ = cv2.imread(occ_path, cv2.IMREAD_GRAYSCALE)
    
    evals = me.evaluate_over_all(img_temp, gt*4, occ, occlusions_counted_in_errors = False)
    average[exp_id]+=evals[2]
    print("{2}:Bad4 for scene {0}: {1:.4f}".format(scene, evals[2], exp_id))

for k, v in average.items():
    print("{0}: {1}".format(k,v/15))
```
