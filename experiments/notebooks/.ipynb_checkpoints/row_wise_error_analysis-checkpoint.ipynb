### Is there a correlation between STD within one line and error rate?

```python
from skimage.filters import sobel_h
img_to_analyse = to_experiment[0]
bad1 =me.rowwise_bad(img_to_analyse, gt, occ, threshold=0, occlusions_counted_in_errors=True)
std = np.std(sobel_h(img_to_analyse), axis=1)
np.corrcoef(bad1, std)
```

### The Pearson correlation suggests otherwise

### Perhaps entropy or gini impurity?

```python
def calculate_entropy(img_row):
    unique, counts = np.unique(img_row, return_counts=True)
    probs = counts/len(img_row)
    return -np.sum(probs*np.log2(probs))
entropies = np.array([calculate_entropy(row) for row in sobel_h(img_to_analyse)])
bad0 =me.rowwise_bad(img_to_analyse, gt, occ, threshold=0, occlusions_counted_in_errors=True)

np.corrcoef(entropies, bad0)
```

```python
def calculate_gini(img_row):
    unique, counts = np.unique(img_row, return_counts=True)
    probs = counts/len(img_row)
    return np.sum(probs*(1-probs))

ginis = np.array([calculate_gini(row) for row in sobel_h(img_to_analyse)])
np.corrcoef(ginis, bad0)
```

### The Pearson coefficient suggests a moderate correlation between error and gini impurity and entropy based on one image. How about all loaded images?

```python
pearsons = []
for img in to_experiment:
    ginis = np.array([calculate_gini(row) for row in img])
    bad0 =me.rowwise_bad(img, gt, occ, threshold=0, occlusions_counted_in_errors=True)
    temp = np.corrcoef(ginis, bad0)
    pearsons.append(temp)
    print(temp)
```

## How about row-wise absolute error?

```python
pearsons_abs = []
for img in to_experiment:
    ginis = np.array([calculate_gini(row) for row in img])
    abs_err =me.rowwise_abs(img, gt, occ, threshold=0, occlusions_counted_in_errors=True)
    temp = np.corrcoef(ginis, abs_err)
    pearsons_abs.append(abs_err)
    print(temp)
```

```python
from skimage.filters import laplace, median, meijering, sato, frangi, hessian, prewitt, sobel, scharr
all_filters = [laplace, median, meijering, sato, frangi, hessian, prewitt, sobel, scharr]

pearsons = []
max = 0
for img in top_teddies:
    for current_filter in all_filters:
        ginis = np.array([calculate_gini(row) for row in current_filter(img)])
        bad0 =me.rowwise_bad(img, gt, occ, threshold=0, occlusions_counted_in_errors=True)
        temp = np.corrcoef(ginis, bad0)
        pearsons.append(temp)
        maximum = max(maximum, abs(temp[0][1]))
        print("Maximum value with filter ({1}):".format(maximum, current_filter.__name__)
```
